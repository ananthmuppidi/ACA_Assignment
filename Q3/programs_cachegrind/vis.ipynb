```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

l1_cache_sizes = [32, 64, 128]
l2_cache_sizes = [64, 128, 256]
miss_ratios = np.array([
    [0.040127, 0.036080, 0.034278],
    [0.0231232, 0.021238, 0.027086],
    [0.0132444, 0.009128, 0.010007]
])

l1_flat = np.tile(l1_cache_sizes, len(l2_cache_sizes))
l2_flat = np.repeat(l2_cache_sizes, len(l1_cache_sizes))
miss_ratios_flat = miss_ratios.flatten()


min_index = np.argmin(miss_ratios_flat)
lowest_miss_ratio = miss_ratios_flat[min_index]
lowest_l1_size = l1_flat[min_index]
lowest_l2_size = l2_flat[min_index]

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(l1_flat, l2_flat, miss_ratios_flat, c=miss_ratios_flat, cmap='viridis', marker='o')


ax.scatter([lowest_l1_size], [lowest_l2_size], [lowest_miss_ratio], c='red', marker='o', s=50, label='Lowest Miss Ratio')


ax.set_xlabel('L1 Cache Size (KB)')
ax.set_ylabel('L2 Cache Size (KB)')
ax.set_zlabel('Miss Ratio')
ax.legend()

# Adding a color bar to represent the miss ratios
plt.colorbar(sc, label='Miss Ratio')

# Show the plot
plt.show()

# Print the point with the lowest miss ratio
print(f"Lowest Miss Ratio: {lowest_miss_ratio}")
print(f"L1 Cache Size (KB): {lowest_l1_size}")
print(f"L2 Cache Size (KB): {lowest_l2_size}")
```
